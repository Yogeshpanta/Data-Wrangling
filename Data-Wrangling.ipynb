## Bivariate Analysis
### is there a chance of drivers under influence of drug or alcohol results in severity or not? (cross_tab function)

```python
underinfl_severity_df = pd.crosstab(df['SEVERITYDESC'],df['UNDERINFL'])
underinfl_severity_df = underinfl_severity_df.drop(index='Unknown')
#underinfl_severity_df = underinfl_severity_df.drop(columns='Unknown')
underinfl_severity_df
```

```python
# hypothesis testing (chisquare test for independence)
# H0: There is no relationship between underinfl and severity
# H1: There is a strong relationship between underinfl and severity
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(underinfl_severity_df)
print(dof)
print(p)
```

Result : there is a strong relation between severitydesc and underinfl (p<0.05)

```python
#group barplot
df1= df[df['UNDERINFL'] != 'Unknown']
df1 = df1[df1['SEVERITYDESC'] != 'Unknown']
severity_underinfl = df1[['SEVERITYDESC','UNDERINFL']].value_counts().reset_index()
severity_underinfl
```

```python
from scipy.stats import kstest,shapiro
statistic,pvalue = kstest(df['PERSONCOUNT'],'norm')   # n>100
print('p-value',pvalue)
```

Accept H1 (dont follow gaussian ,no normal distribution) 


```python
from scipy.stats import ttest_ind

group_0 = df[df['PERSONCOUNT'] == 0]['PERSONCOUNT']
group_1 = df[df['PERSONCOUNT'] == 1]['PERSONCOUNT']

# Perform independent t-test
statistic, pvalue = ttest_ind(group_0, group_1)

# Print the p-value
print('p-value:', pvalue)
```

we reject null hypothesis and conclude that there is indeed a signifiacant difference between the two groups being compared.
